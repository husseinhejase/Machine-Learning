### Activation functions
Why does a neural network need a non-linear activation function?
Turns out that a neural network needs non-linear activation functions to compute interesting functions and produce a non-linear decision boundary.

**Sigmoid:**
$$f(z) = \frac{1}{1+e^{-z}}$$

**tanh:**a
$$f(z) = \frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$$

**ReLU:**
$$f(z) = 
    \begin{cases}
      z & if z > 0\\
      0 & else\\
    \end{cases}$$

**Leaky ReLU:**
$$f(z) = 
    \begin{cases}
      z & if z > 0\\
      0.01z & else\\
    \end{cases}$$
    
A few tips on the usage of these functions:
- tanh is strictly superior to sigmoid. tanh almost always works better than sigmoid because it does a better job in centering the data ($\mu = 0$). tanh is a shifted version of the sigmoid.
- You could use a sigmoid function in the output layer if you are doing binary classification.
- For sigmoid and tanh, when z is very small or large, the gradients become small slowing down the learning process and gradient descent.
- An alternative efficient (learning is quite fast!) is using ReLU or leaky ReLU. In practice, you should have enough hidden units with z > 0. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(z):
    """
    Sigmoid function
    """
    return 1 / (1 + np.exp(-z))

def tanh(z):
    """
    Hyperbolic tangent function
    """
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

def relu(z):
    """
    Rectified Linear Unit (ReLU) function
    """
    return np.maximum(z, 0)

def leaky_relu(z, param):
    """
    Leaky Rectified Linear Unit (ReLU) function
    """
    return np.maximum(z, param*z)

def plot_fxn(z, fz, name):
    plt.plot(z, fz, '-', color='r')
    plt.xlabel('z', fontsize=18)
    plt.ylabel('f(z)', fontsize=18)
    plt.title(name, fontsize=22)
```

```python
z = np.array(np.arange(-10,10,0.1))
```

```python
fz = sigmoid(z)
plot_fxn(z, fz, 'Sigmoid function')
```

```python
fz = tanh(z)
plot_fxn(z, fz, 'tanh function')
```

```python
fz = relu(z)
plot_fxn(z, fz, 'ReLU function')
```

```python
param = 0.1
fz = leaky_relu(z, param)
plot_fxn(z, fz, 'Leaky ReLU function')
```
